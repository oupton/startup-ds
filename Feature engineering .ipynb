```python
import pandas as pd
import numpy as np
import os
import ast
#from joblib import Parallel, delayed
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import ast
```

```python
movie_df = pd.read_csv('data/combined_metadata.csv')
movie_df = movie_df.drop(['Unnamed: 0'], axis=1)
movie_df
```

```python
movie_df['genres'] = movie_df.apply(lambda row: ast.literal_eval(row.genres), axis=1)
movie_df['production_countries'] = movie_df.apply(lambda row: ast.literal_eval(row.production_countries), axis=1)
movie_df['spoken_languages'] = movie_df.apply(lambda row: ast.literal_eval(row.spoken_languages), axis=1)
```

```python
def english (row):
    if row.spoken_languages:
        for lang in row.spoken_languages:
            if lang['name'] == 'English':
                return 1
    return 0
            
def numlang (row):
    if row.spoken_languages:
        return len(row.spoken_languages)
    
def listlangs (row):
    if row.spoken_languages:
        for lang in row.spoken_languages:
            if lang['name'] != 'English':
                print lang['name']
```

```python
movie_df.apply(listlangs, axis=1)
```

```python
#movie_df['English'] = movie_df.apply(english, axis=1)
movie_df['numlang'] = movie_df.apply(numlang, axis=1)
```

```python
movie_df
```

```python
one_hot = pd.get_dummies(movie_df['original_language'])
movie_df['original_language'].value_counts()
```

```python
# remove obscure languages to avoid memorization
for col in one_hot:
    if sum(one_hot[col]) < 10:
        one_hot = one_hot.drop([col], 1)
```

```python
one_hot
```

```python
movie_df = movie_df.join(one_hot)
```

```python
movie_df = movie_df.drop(['original_language'],1)
movie_df = movie_df.drop(['spoken_languages'],1)
```
