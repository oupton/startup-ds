```python
import pandas as pd
import numpy as np
import ast
import os
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(os.path.join('data', 'alldata_date_features.csv'))
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
genre_set = set()
for genres in df.genres:
    genres = ast.literal_eval(genres)
    for genre in genres:
        genre_set.add(genre['name'])
```

```python
genre_set = list(genre_set)
for genre in genre_set:
    df[genre] = np.zeros(len(df)).astype('int')
```

```python
df.head()
```

```python
for i in range(len(df)):
    genres = ast.literal_eval(df.genres[i])
    for genre in genres:
        df.loc[i, genre['name']] = 1
```

```python
df.head()
```

```python
df = df.drop('genres', axis=1)
```

```python
for genre in genre_set:
    print(df[genre].value_counts())
    print('\n\n')
```

```python
final_df = df[df['TV Movie'] == 0]
```
